```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

gamma = np.random.gamma(5,1,100)
```

```python
plt.hist(gamma)
plt.show()
```

```python
mean = gamma.mean()
sd = gamma.std()

plt.hist(gamma)
plt.axvline(mean, color='black', linestyle='solid')
plt.axvline(mean+sd, color='black', linestyle='dashed')
plt.axvline(mean-sd, color='black', linestyle='dashed')
plt.show()
```

Mean may not be the best descriptor of central tendency for this dataset as the data cleary demonstrates a right skeweness. However, the data may be normal enough for the mean to give a general idea of where the data is centered. The standard deviation can still be useful for this data to describe variability.

```python
rand1 = np.random.normal(5, 0.5, 100)
rand2 = np.random.normal(10, 1, 100)
rand3 = rand1 + rand2

plt.hist(rand3)
plt.show()
```

```python
u_rand3 = rand3.mean()
sd_rand3 = rand3.std()

plt.hist(rand3)
plt.axvline(u_rand3, color='black', linestyle='solid')
plt.axvline(u_rand3+sd_rand3, color='black', linestyle='dashed')
plt.axvline(u_rand3-sd_rand3, color='black', linestyle='dashed')
plt.show()
```

```python
rand3.sort()
rand1.sort()

plt.plot(rand3, rand1, 'o')
plt.show()
```

The addition of rand1 and rand2 may have created a semi-multimodal distribution that is not effectively summarized simply by a mean and standard deviation. While these statistics may still be valuable in describing the data, the unique characteristics of the new distribution should be taken in consideration. QQ plots of rand3 reveal a relatively normal distribution; indicating that mean and standard deviation are valuable summary statistics.
